```python
import pandas as pd
```

```python
crosswalk = pd.read_csv(r'C:\Users\ywang\Documents\GitHub\onboard-surveys\make-uniform\production\canonical_route_crosswalk.csv')
print('canonical_crosswalk has {} rows'.format(crosswalk.shape[0]))
```

#### Create 'GEOCODE' rows for all ACE, SMART, and CC stations

```python
# All ACE stations

ace_data = pd.read_csv(r'M:\Data\OnBoard\Data and Reports\ACE\2019\As CSV\ACE19_Final Data_AddCols_RecodeRoute_NO POUND OR SINGLE QUOTE.csv')

ace_station_dict = {1: 'STOCKTON', 
                    2: 'LATHROP / MANTECA',
                    3: 'TRACY',
                    4: 'VASCO ROAD', 
                    5: 'LIVERMORE',
                    6: 'PLEASANTON', 
                    7: 'FREMONT', 
                    8: 'GREAT AMERICA', 
                    9: 'SANTA CLARA', 
                    10: 'SAN JOSE'}

for i in ['ACE_Boarding', 'ACE_Alighting']:
    ace_data[i+'_station'] = ace_data[i].map(ace_station_dict)

ace_stations = list(ace_data.ACE_Boarding_station.unique())
ace_stations.extend(list(ace_data.ACE_Alighting_station.unique()))
ace_stations = list(set(ace_stations))

print('ACE has {} stations'.format(len(ace_stations)))
print(ace_stations)

# all possible combinations of stations
all_comb_ace = []
for i in ace_stations:
    for j in ace_stations:
        comb = i + '&&&' + j
        all_comb_ace.append(comb)
# print(all_comb_ace)
print('ACE has {} unique combinations of stations'.format(len(all_comb_ace)))

# create "GEOCODE" dataframe
GEOCODE_add_ace = pd.DataFrame(columns = list(crosswalk))
GEOCODE_add_ace['survey_name'] = all_comb_ace
GEOCODE_add_ace['canonical_name'] = GEOCODE_add_ace['survey_name']
GEOCODE_add_ace['canonical_operator'] = 'ACE'
GEOCODE_add_ace['operator_detail'] = GEOCODE_add_ace['canonical_operator']
GEOCODE_add_ace['technology'] = 'commuter rail'
GEOCODE_add_ace['technology_detail'] = GEOCODE_add_ace['technology']
GEOCODE_add_ace['survey'] = 'GEOCODE'
GEOCODE_add_ace['survey_year'] = 2018

print(GEOCODE_add_ace.shape)
display(GEOCODE_add_ace)
```

```python
# All SMART stations

smart_data = pd.read_csv(r'M:\Data\OnBoard\Data and Reports\SMART\As CSV\SMART Standardized Final Data_addRouteCols_NO POUND NO SINGLE QUOTE.csv')

smart_stations = list(smart_data.Board_Station.unique())
smart_stations.extend(list(smart_data.Alight_station.unique()))

# SMART opened two new stations in 2019, Novato Downtown and Larkspur,
# which should not be added to the crosswalk since the survey is conducted in 2018

smart_stations = list(set(smart_stations))
print('Smart has {} stations'.format(len(smart_stations)))
print(smart_stations)

# all possible combinations of stations
all_comb_smart = []
for i in smart_stations:
    for j in smart_stations:
        comb = i + '&&&' + j
        all_comb_smart.append(comb)
# print(all_comb_smart)
print('SMART has {} unique combinations of stations'.format(len(all_comb_smart)))

# create "GEOCODE" dataframe
GEOCODE_add_smart = pd.DataFrame(columns = list(crosswalk))
GEOCODE_add_smart['survey_name'] = all_comb_smart
GEOCODE_add_smart['canonical_name'] = GEOCODE_add_smart['survey_name']
GEOCODE_add_smart['canonical_operator'] = 'SMART'
GEOCODE_add_smart['operator_detail'] = GEOCODE_add_smart['canonical_operator']
GEOCODE_add_smart['technology'] = 'commuter rail'
GEOCODE_add_smart['technology_detail'] = GEOCODE_add_smart['technology']
GEOCODE_add_smart['survey'] = 'GEOCODE'
GEOCODE_add_smart['survey_year'] = 2018

print(GEOCODE_add_smart.shape)
display(GEOCODE_add_smart)
```

```python
# All Capitol Corridor stations

cc_data = pd.read_csv(r'M:\Data\OnBoard\Data and Reports\Capitol Corridor\OD Survey 2019\As CSV\CAPCO19 Data-For MTC_NO POUND OR SINGLE QUOTE.csv')

cc_stations = list(cc_data.Q1A.dropna().unique())
cc_stations.extend(list(cc_data.Q1B.dropna().unique()))

cc_stations = list(set(cc_stations))

print('Capitol Corridor has {} stations'.format(len(cc_stations)))
print(sorted(cc_stations))

# all possible combinations of stations
all_comb_cc = []
for i in cc_stations:
    for j in cc_stations:
        comb = i + '&&&' + j
        all_comb_cc.append(comb)
# print(all_comb_cc)
print('Capitol Corridor has {} unique combinations of stations'.format(len(all_comb_cc)))

# create "GEOCODE" dataframe
GEOCODE_add_cc = pd.DataFrame(columns = list(crosswalk))
GEOCODE_add_cc['survey_name'] = all_comb_cc
GEOCODE_add_cc['canonical_name'] = GEOCODE_add_cc['survey_name']
GEOCODE_add_cc['canonical_operator'] = 'Capitol Corridor'
GEOCODE_add_cc['operator_detail'] = GEOCODE_add_cc['canonical_operator']
GEOCODE_add_cc['technology'] = 'commuter rail'
GEOCODE_add_cc['technology_detail'] = GEOCODE_add_cc['technology']
GEOCODE_add_cc['survey'] = 'GEOCODE'
GEOCODE_add_cc['survey_year'] = 2018

print(GEOCODE_add_cc.shape)
display(GEOCODE_add_cc)
```

#### Create survey route rows for ACE, SMART, and CC

```python
# ACE survey routes
ace_data['survey_route_for_crosswalk'] = ace_data['ACE_Boarding_station'] + '&&&' + ace_data['ACE_Alighting_station']
ace_routes = list(ace_data.survey_route_for_crosswalk.unique())
print('ACE survey has {} unique survey_routes'.format(len(ace_routes)))
# print(ace_routes)

# create rows
route_add_ace = pd.DataFrame(columns = list(crosswalk))
route_add_ace['survey_name'] = ace_routes
route_add_ace['canonical_name'] = route_add_ace['survey_name']
route_add_ace['canonical_operator'] = 'ACE'
route_add_ace['operator_detail'] = route_add_ace['canonical_operator']
route_add_ace['technology'] = 'commuter rail'
route_add_ace['technology_detail'] = route_add_ace['technology']
route_add_ace['survey'] = 'ACE'
route_add_ace['survey_year'] = 2019

print(route_add_ace.shape)
display(route_add_ace)
```

```python
# SMART survey routes
smart_data['survey_route_for_crosswalk'] = smart_data['Board_Station'] + '&&&' + smart_data['Alight_station']
smart_routes = list(smart_data.survey_route_for_crosswalk.unique())
print('SMART survey has {} unique survey_routes'.format(len(smart_routes)))
# print(smart_routes)

# create rows
route_add_smart = pd.DataFrame(columns = list(crosswalk))
route_add_smart['survey_name'] = smart_routes
route_add_smart['canonical_name'] = route_add_smart['survey_name']
route_add_smart['canonical_operator'] = 'SMART'
route_add_smart['operator_detail'] = route_add_smart['canonical_operator']
route_add_smart['technology'] = 'commuter rail'
route_add_smart['technology_detail'] = route_add_smart['technology']
route_add_smart['survey'] = 'Sonoma-Marin Area Rail Transit'
route_add_smart['survey_year'] = 2018

print(route_add_smart.shape)
display(route_add_smart)
```

```python
# Capitol Corridor survey routes
cc_data['survey_route_for_crosswalk'] = cc_data['Q1A'] + '&&&' + cc_data['Q1B']
cc_routes = list(cc_data.survey_route_for_crosswalk.unique())
print('Capitol Corridor survey has {} unique survey_routes'.format(len(cc_routes)))
# print(cc_routes)

# create rows
route_add_cc = pd.DataFrame(columns = list(crosswalk))
route_add_cc['survey_name'] = cc_routes
route_add_cc['canonical_name'] = route_add_cc['survey_name']
route_add_cc['canonical_operator'] = 'Capitol Corridor'
route_add_cc['operator_detail'] = route_add_cc['canonical_operator']
route_add_cc['technology'] = 'commuter rail'
route_add_cc['technology_detail'] = route_add_cc['technology']
route_add_cc['survey'] = 'Capitol Corridor'
route_add_cc['survey_year'] = 2019

print(route_add_cc.shape)
display(route_add_cc)
```

#### combine all rows to be added to canonical_crosswalk

```python
add_crosswalk = pd.concat([GEOCODE_add_ace, GEOCODE_add_smart, GEOCODE_add_cc,
                           route_add_ace, route_add_smart, route_add_cc])

print('{} rows to be added to canonical_crosswalk'.format(add_crosswalk.shape[0]))
```

```python
# add to crosswalk and check for duplicates
# this step runs only once

crosswalk_new = pd.concat([crosswalk, add_crosswalk])
print('the new crosswalk has {} rows'.format(crosswalk_new.shape[0]))

crosswalk_new_dedup = crosswalk_new.drop_duplicates()
print('after dropping duplicates, the new crosswalk has {} rows'.format(crosswalk_new_dedup.shape[0]))
```
